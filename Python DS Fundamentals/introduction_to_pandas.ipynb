```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos un diccionario que contiene la data que vamos a trabajar 
carros_dataset= {
    'Marca' : ['Tesla','BMW', 'Hyundai'], 
    'Modelo' : ['MODEL S', 'i3', 'IONIQ 5'],
    'Rango' : [391, 335, 369],
    'Aceleracion' : [3.7, 3.3, 4.5]
}

# Creamos un DataFrame de pandas con el diccionario que creamos antes
df_carros = pd.DataFrame(carros_dataset)
```

```python
# Mostramos el DataFrame
df_carros
```

```python
# Revisamos el tipo de dato de el DF
type(df_carros)
```

```python
# Revisamos el tipo de dato de la columna, siempre son Series de pandas 
type(df_carros['Marca'])
```

```python
# Realizamos un filtro para obtener solo las columnas que mencionamos 
df_carros[['Marca','Modelo']]
```

```python
# Creamos una columna llamada impulsor que contrendra un array de numpy con numeros Float
df_carros['Turbo']=np.array([.4, .6, .3])
df_carros
```

```python
# Creamos la columna Acel_turbo que es resultado de multiplicar las columnas Aceleracion y Turbo
df_carros['Acel_turbo'] = df_carros['Aceleracion'] * df_carros['Turbo']
df_carros
```

```python
# Eliminamos una columna con el metodo .drop() y el parametro inplace=True, para que modifque el DF original

df_carros.drop(columns=['Acel_turbo'], inplace=True)
df_carros
```

```python
# Renombrar el titulo de las columnas con el metodo .rename()

df_carros.rename(columns={'Marca': 'marca'}, inplace=True)
df_carros
```

```python
# convertimos el titulo de todas las columnas en minusculas(lowercase)

df_carros.columns = df_carros.columns.str.lower()
df_carros
```

```python
ventas.head()
```

```python
# Si Filtramos de esta forma obtendremos una lista de booleanos marcando False las que no cumplen la condicion
ventas['Frozen'] > 5000
# Si filtramos asi obtenemos el DF Filtrado por la condicion
ventas[ventas['Frozen'] > 5000]
```

```python
# Filtramos por 2 o mas condiciones condiciones instanciadas entre ()
ventas[(ventas['Frozen'] > 5000) & (ventas['Milk'] < 600)]
```

```python
# Filtro por 2 condiciones 
ventas[(ventas['Frozen'] > 2000) & (ventas['Channel'].isin([1, 2]))]
```

LOC & ILOC


```python
# Para buscar exactamente un valor con .loc se coloca [ index, columna ] 
ventas.loc[2, 'Milk']
```

```python
# Seleccionamos solamente la fila con el index 4 
ventas.loc[4]
```

```python
# para buscar un valor con iloc se colocan solo numeros [ index, N_columna ]
ventas.iloc[2,3]
```

```python
# Revisamos cantidad de filas y columnas
ventas.shape
```

```python
# Iteramos sobre la columna Milk para cambiar los valore por numero de index
for index in range(440):
    ventas.loc[index, 'Milk'] = index

ventas    
```
